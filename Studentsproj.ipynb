# Drop Unnamed Column

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

# Change weekly study hours column

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("5 - 10","5-10")
df.head()
```

Gender Distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x= "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we analysed that:the number of females is more than the number of males
```

```python
gb=df.groupby("ParentEduc").aggregate({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.title("Relationship btwn parents education and students score")
plt.show()
```

```python
#from the above chart we have concluded that the education of their parents have a good impact on their scores
```

```python
gb1=df.groupby("ParentMaritalStatus").aggregate({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.title("Relationship btwn parents marital status and students score")
plt.show()
```

```python
#from the above chart we have concluded that the marital status of their parents have no impact/negligible impact on their scores
```

# Distribution of Ethnic groups

```python
grpA = df.loc[(df["EthnicGroup"] == "group A")].count()
grpB = df.loc[(df["EthnicGroup"] == "group B")].count()
grpC = df.loc[(df["EthnicGroup"] == "group C")].count()
grpD = df.loc[(df["EthnicGroup"] == "group D")].count()
grpE = df.loc[(df["EthnicGroup"] == "group E")].count()

l=["group A","group B","group C","group D","group E"]
mlist=[grpA["EthnicGroup"],grpB["EthnicGroup"],grpC["EthnicGroup"],grpD["EthnicGroup"],grpE["EthnicGroup"]]


plt.title("Distribution of Ethnic Groups")
colors=['Pink','Lavender','LightBlue','LightGrey','LightGreen']
plt.axis('equal')
plt.pie(mlist,labels=l,autopct= "%1.2f%%",colors=colors)
```
